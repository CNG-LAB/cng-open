```python
from brainspace.plotting import plot_hemispheres
from brainspace.datasets import load_parcellation
from brainspace.utils.parcellation import map_to_labels 
from brainspace.datasets import load_conte69
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from nilearn import plotting
import ptitprince as pt
import nibabel as nb
import seaborn as sns
from matplotlib.patches import Rectangle
import matplotlib.colors as colors
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from scipy import stats

plt.rcParams["pdf.fonttype"]=42  #export as editable text
plt.rcParams["ps.fonttype"]=42
plt.rcParams["svg.fonttype"]="none" #export as editable text

plt.rc("font",size=10)
plt.rc("axes", labelsize=22)
plt.rc("xtick", labelsize=13)
plt.rc("ytick", labelsize=13)
plt.rcParams['axes.linewidth'] = 0.8
screenshot = True
```

# Proof ground truth by plotting connectivity profiles of thomas nuclei

### struc conn

```python
#AV
nucleus_label=2  #VLP=6 12=MD    ####8=Pulvinar 2=AV  4=VA  7=VPL  10=MGN  #VPL
nucleus_l=np.load("/Project1_thalamus_gradients/data/structural_connectivity/projections_thomasnuclei/sc_nucleus_{}_l.npy".format(nucleus_label))
nucleus_r=np.load("/Project1_thalamus_gradients/data/structural_connectivity/projections_thomasnuclei/sc_nucleus_{}_r.npy".format(nucleus_label))

nucleus_conn=np.concatenate((nucleus_l.flatten(), nucleus_r.flatten()))

surf_lh,surf_rh=load_conte69()
labeling=load_parcellation("schaefer",scale=200,join=True)
mask= labeling !=0
fc_nucleus=map_to_labels(nucleus_conn,labeling,mask,fill=np.nan)

plot_hemispheres(surf_lh, surf_rh, array_name= fc_nucleus,
                 cmap="viridis",color_bar=True,
                 background=(1,1,1),size=(1200,200),zoom=1.5,
                 color_range=(0,0.7),
                 embed_nb=True,
                 interactive=False,
                 screenshot=screenshot, 
                 transparent_bg=True,
                 filename = "/Project1_thalamus_gradients/plots/Supp_thomas_nuclei_projection/sc_AV.png")   #change name
```
