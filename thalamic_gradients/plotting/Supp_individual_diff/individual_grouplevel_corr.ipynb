```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as colors

plt.rcParams["pdf.fonttype"]=42  #export as editable text
plt.rcParams["ps.fonttype"]=42
plt.rcParams["svg.fonttype"]="none" #export as editable text
plt.rcParams["font.family"]="Arial"
```

```python
# import correlation arrays

sc_1_lh_aligned=np.load("/Project1_thalamus_gradients/data/structural_connectivity/parcels_200/individual_gradients/corr_individual_group_gradients_1_lh_aligned.npy")
sc_2_lh_aligned=np.load("/Project1_thalamus_gradients/data/structural_connectivity/parcels_200/individual_gradients/corr_individual_group_gradients_2_lh_aligned.npy")

sc_1_rh_aligned=np.load("/Project1_thalamus_gradients/data/structural_connectivity/parcels_200/individual_gradients/corr_individual_group_gradients_1_rh_aligned.npy")
sc_2_rh_aligned=np.load("/Project1_thalamus_gradients/data/structural_connectivity/parcels_200/individual_gradients/corr_individual_group_gradients_2_rh_aligned.npy")

fc_1_lh_aligned=np.load("/Project1_thalamus_gradients/data/functional_connectivity/individual_gradients/corr_individual_group_gradients_1_lh_aligned.npy")
fc_2_lh_aligned=np.load("/Project1_thalamus_gradients/data/functional_connectivity/individual_gradients/corr_individual_group_gradients_2_lh_aligned.npy")

fc_1_rh_aligned=np.load("/Project1_thalamus_gradients/data/functional_connectivity/individual_gradients/corr_individual_group_gradients_1_rh_aligned.npy")
fc_2_rh_aligned=np.load("/Project1_thalamus_gradients/data/functional_connectivity/individual_gradients/corr_individual_group_gradients_2_rh_aligned.npy")


T1q_lh=np.load("/Project1_thalamus_gradients/data/structural_covariance/Revision_individual_qT1maps/corr_individual_group_qT1maps_lh.npy")
T1q_rh=np.load("/Project1_thalamus_gradients/data/structural_covariance/Revision_individual_qT1maps/corr_individual_group_qT1maps_rh.npy")
```

```python

# Create a DataFrame
data = pd.DataFrame({"G1sc LH": sc_1_lh_aligned, "G1sc RH": sc_1_rh_aligned, 
                     "G2sc LH": sc_2_lh_aligned, "G2sc RH": sc_2_rh_aligned,
                     "G1fc LH": fc_1_lh_aligned, "G1fc RH": fc_1_rh_aligned, 
                     "G2fc LH": fc_2_lh_aligned, "G2fc RH": fc_2_rh_aligned,
                     "qT1 LH": T1q_lh, "qT1 RH" : T1q_rh, })
data
```

```python
# Create the violin plot
plt.figure(figsize=(9, 5))

#create colors for the 3 modalitites (matching with fig.3)
mako_colors=sns.color_palette("mako",8)
mako_color=mako_colors[3]
cividis_colors=sns.color_palette("cividis",8)
cividis_colors=cividis_colors[6]
qT1_cmap=sns.cubehelix_palette(start=2, rot=1, dark=0.3, light=0.9, reverse=True, n_colors=8)
qT1_cmap=qT1_cmap[2]

combined_palette=[mako_color]*4  + [cividis_colors]*4 + [qT1_cmap]*2

#plot violin
sns.violinplot(data=data, palette=combined_palette, width=1.4, linewidth=1.2)
plt.yticks(fontsize=14)
plt.xticks(rotation=45, fontsize=14)
plt.savefig("/Project1_thalamus_gradients/plots/Revision_Supplements/corr_individualmap_with_grouplevel.svg", transparent=True, format='svg', dpi=600)
```
